## Q7. Sum of weights

```python
df_laptop_Innjoo = df[df['Brand']=='Innjoo']
df_laptop_Innjoo
```

```python
df_laptop_Innjoo = df_laptop_Innjoo[['RAM', 'Storage', 'Screen']]
```

```python
df_laptop_Innjoo
```

```python
X= df_laptop_Innjoo.to_numpy()
```

```python
X
```

```python
XT = X.T
```

```python
XT
```

```python
XTX = np.matmul(XT,X)
```

```python
XTX
```

```python
INTX=inv(XTX)
```

```python
INTX
```

```python
y= np.array([1100, 1300, 800, 900, 1000, 1100])
```

```python
yt = y.T
```

```python
r1 = np.matmul(INTX, XT)
r1
```

```python
w=np.matmul(r1, yt)
```
